### **Metric Evaluation**

Dalam proyek ini, evaluasi kinerja model tidak hanya berfokus pada akurasi secara keseluruhan, melainkan lebih menitikberatkan pada dampak bisnis yang timbul dari kesalahan prediksi. Mengingat tujuan utama bank adalah mengidentifikasi nasabah dengan potensi tinggi untuk melakukan deposito, maka konsekuensi finansial dari berbagai jenis kesalahan klasifikasi perlu diperhitungkan secara berbeda, khususnya pada dua jenis kesalahan utama:

- FP:
    - Model memprediksi nasabah akan deposit, padahal sebenarnya tidak.
    - Jika nasabah diprediksi akan deposit, maka nasabah akan diberikan kampanye pemasaran deposito. Potensi kerugiannya adalah biaya kampanye pemasaran deposito terhadap nasabah tersebut.
    - Asumsi cost kampanye pemasaran deposito untuk 1 nasabah sebesar 100 ribu.
    - Asumsi cost ini diambil dari referensi biaya call center dengan tenaga profesional di Indonesia untuk 15 menit sebesar 80000, biaya materi digital sebesar 25000, dan biaya telemarketing dan follow up sebesar 40000. Sehingga totalnya 145000 dan dibulatkan menjadi 150000.
- FN:
    - Model memprediksi nasabah tidak akan deposit, padahal sebenarnya tertarik/akan deposit.
    - jika nasabah diprediksi tidak akan deposito, maka tidak akan memberikan kampanye pemasaran deposito yang dapat menyebabkan kehilangan pendapatan deposito dari nasabah tersebut.
    - Asumsi cost kehilangan pendapatan deposito dari 1 nasabah sebesar 1,2 juta.
    - Asumsi cost ini diambil dari median nilai deposit orang indonesia sebesar 20 juta per nasabah dengan bunga yang diberikan oleh bank sebesar 4% (800000). Uang deposito disalurkan bank untuk peminjaman dengan bunga 10% (2 juta). Sehingga keuntungan yang seharusnya didapatkan bank sebesar 1,2 juta untuk 1 nasabah.

False Negative (FN) jauh lebih mahal, karena berisiko menyebabkan kehilangan nasabah yang akan deposito.
Karena FN memiliki dampak finansial yang lebih besar, maka fokus model adalah menghindari kesalahan FN sebanyak mungkin namun masih mempertimbangkan nilai FP.

Maka metric yang digunakan adalah **F2-Score**.

## **5. Preprocessing**

Dalam tahap preprocessing, dilakukan beberapa transformasi untuk mempersiapkan data sebelum dimasukkan ke dalam model machine learning. Beberapa fitur kategorikal dikodekan sesuai dengan karakteristiknya. Fitur-fitur seperti 'job', 'housing', 'loan', 'contact', 'poutcome', dan 'month' dikodekan menggunakan One-Hot Encoding karena tidak memiliki tingkatan dan bersifat nominal.

Untuk fitur numerik seperti 'age', 'balance', 'campaign', dan 'pdays', digunakan Robust Scaler. Pemilihan scaler ini bertujuan untuk mengurangi pengaruh nilai ekstrem (outliers) pada proses pelatihan model, karena Robust Scaler melakukan penskalaan berdasarkan interquartile range (IQR) dan bukan mean dan standar deviasi seperti StandardScaler.

Langkah-langkah preprocessing ini dilakukan agar model dapat belajar dari data dalam format numerik yang konsisten, terstandardisasi, dan optimal untuk performa prediktif.
- Encoding:
    - OneHot: 'job', 'housing', 'loan', 'contact', 'poutcome', 'month'
- Scaler:
    - Robust : 'age', 'balance', 'campaign', 'pdays'

```python
transformer = ColumnTransformer([
    ('onehot',OneHotEncoder(drop='first'),['job', 'housing', 'loan', 'contact', 'poutcome','month']),
    ('scaler',MinMaxScaler(),['age', 'balance', 'campaign', 'pdays'])
])
```

## **6. Cross Validation**

Pada tahap ini akan dilakukan modeling tanpa dilakukan hyperparameter tuning pada beberapa opsi model machine learning yaitu:
- LogisticRegression
- KNeighborsClassifier
- DecisionTreeClassifier
- RandomForestClassifier
- AdaBoostClassifier
- GradientBoostingClassifier
- LGBMClassifier
- XGBClassifier

Model yang terbaik akan dipilih berdasarkan skor f2-score dan akan digunakan untuk memprediksi nilai pada data X_test. Hasil prediksi tersebut akan menjadi benchmark nilai model sebelum melakukan hyperparameter tuning pada tahapan berikutnya.

```python
# define algo
logistic = LogisticRegression(random_state=0)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=0)
rf = RandomForestClassifier(random_state=0)
adaboost = AdaBoostClassifier(random_state=0,estimator=DecisionTreeClassifier(random_state=0,max_depth=1))
gradboost = GradientBoostingClassifier(random_state=0)
lgbm = LGBMClassifier(random_state=0)
xgboost = XGBClassifier(random_state=0)
```

```python
# Define scorer
f2_scorer = make_scorer(fbeta_score, beta=2)
```

```python
list_model = [logistic, knn, tree, rf, adaboost, gradboost,lgbm,xgboost]

list_f2score_all = []
list_f2score_mean = []
list_f2score_std = []

for model in list_model:
    # pipe_model
    pipe_model = Pipeline([
        ('preprocession', transformer),
        ("modeling", model)
    ])

    # cv
    cv_score = cross_val_score(
        estimator = pipe_model,
        X= X_train,
        y= y_train,
        cv= 5,
        scoring = f2_scorer
    )
    list_f2score_all.append(cv_score.round(2))
    list_f2score_mean.append(cv_score.mean())
    list_f2score_std.append(cv_score.std())
```

```python
df_cv = pd.DataFrame()
df_cv["model"] = list_model
df_cv["f2_scorer"] = list_f2score_all
df_cv["f2_scorer_mean"] = list_f2score_mean
df_cv["f2_scorer_std"] = list_f2score_std
df_cv
```

Terlihat bahwa model **AdaBoostClassifier** memiliki nilai rata-rata evaluasi metrics yang terbaik dan standar devisiasi yang tidak besar menandakan model tersebut memiliki nilai evaluasi metrics yang stabil. Maka dari itu, model ini yang akan digunakan pada hyperparameter tuning dan ditest dengan data test.

## **7. Hyperparameter Tuning**

**AdaBoostClassifier**

Untuk mendapatkan performa model terbaik, dilakukan proses hyperparameter tuning menggunakan RandomizedSearchCV dengan nilai n_iter sebesar 3000. Pendekatan ini mengevaluasi 3000 kombinasi dari semua kombinasi parameter yang telah ditentukan, menggunakan skema validasi silang sebanyak 5 fold dan skor evaluasi berdasarkan f2-score.

Adapun parameter yang dituning meliputi:
- Scaler : Menguji tiga jenis skala untuk memastikan transformasi fitur numerik tidak mengganggu performa model.
- n_estimators: Menentukan jumlah pohon yang akan dibangun dalam boosting.
- learning_rate : Menentukan seberapa besar kontribusi setiap pohon dalam proses boosting.
- estimator__max_depth : Mengontrol kedalaman maksimum setiap pohon.
- estimator__min_samples_split : Menentukan jumlah minimum sampel yang dibutuhkan untuk membagi sebuah node internal.
- estimator__min_samples_leaf : Menentukan jumlah minimum sampel yang harus dimiliki oleh sebuah leaf node.

Proses ini dibungkus dalam sebuah Pipeline untuk memastikan tahapan preprocessing dan pemodelan dilakukan secara konsisten selama proses tuning.

```python
# define hyperparameter ADA
hyperparam = {
    "modeling__n_estimators": [100, 200, 300],
    "modeling__learning_rate": [0.01, 0.1],
    "modeling__estimator__max_depth" : range(1, 10),
    "modeling__estimator__min_samples_split" : range(5, 15),
    "modeling__estimator__min_samples_leaf" : range(5, 15),
    "preprocessing__scaler":[StandardScaler(), MinMaxScaler(), RobustScaler()]
}

# define pipe_model
pipe_model = Pipeline([
    ("preprocessing", transformer),
    ("modeling", adaboost)
])

randomsearch = RandomizedSearchCV(
    estimator= pipe_model,
    param_distributions=hyperparam,
    scoring = f2_scorer,
    n_jobs = -1,
    cv = 5,
    n_iter= 3000,
    random_state= 0
)
```

```python
randomsearch.fit(X_train, y_train)
```

```python
pd.DataFrame(randomsearch.cv_results_).sort_values("rank_test_score").head()
```

Model terbaik yang diperoleh adalah AdaBoostClassifier dengan konfigurasi parameter sebagai berikut:
- Preprocessing: StandardScaler()
- Jumlah estimator (n_estimators): 300
- Learning rate (learning_rate): 0.1
- Parameter estimator dasar (estimator):
    - Minimum sampel untuk melakukan split (min_samples_split): 9
    - Minimum sampel pada daun (min_samples_leaf): 14
    - Maksimum kedalaman pohon (max_depth): 9

Konfigurasi ini menghasilkan performa optimal pada model dalam memprediksi nasabah yang berpotensi melakukan deposito.

Meskipun model telah melalui proses hyperparameter tuning yang komprehensif dan menunjukkan peningkatan performa secara umum, evaluasi terhadap hasil prediksi pada data uji mengindikasikan bahwa penggunaan threshold default (0.5) belum sepenuhnya mencerminkan kinerja optimal dalam konteks metrik F2-score.

Mengingat bahwa F2-score memberikan bobot lebih besar pada FN dibanding FP dan sesuai dengan tujuan analisis yang menekankan pentingnya meminimalkan kesalahan tipe FN (false negative) sehingga dilakukan penyesuaian threshold terhadap probabilitas keluaran model.

Langkah ini dimaksudkan untuk mengkalibrasi kembali batas pengambilan keputusan klasifikasi, dengan tujuan meningkatkan kemampuan model dalam mendeteksi sebanyak mungkin kasus positif yang relevan, tanpa mengabaikan FP secara signifikan.

```python
model=randomsearch.best_estimator_
model.fit(X_train,y_train)

th_value=[i/100 for i in range(1,99)]
score=0
list_score_f2=[]
list_score_recall=[]
list_score_precision=[]
list_score_akurasi=[]

for i in th_value:
    y_prob=model.predict_proba(X_test)[:,1]
    y_pred=np.where(y_prob>i,1,0)
    score_f2=fbeta_score(y_test, y_pred, beta=2)
    score_recall=recall_score(y_test, y_pred)
    score_precision=precision_score(y_test, y_pred)
    score_akurasi=accuracy_score(y_test, y_pred)
    list_score_f2.append(score_f2)
    list_score_recall.append(score_recall)
    list_score_precision.append(score_precision)
    list_score_akurasi.append(score_akurasi)

th_f1 = pd.DataFrame({
    'th_value':th_value,
    "f2_score_proba":list_score_f2,
    "recall_score_proba":list_score_recall,
    "precision_score_proba":list_score_precision,
    "akurasi_score_proba":list_score_akurasi
})
```

Grafik menggambarkan perubahan empat metrik evaluasi, yaitu: Recall (hijau), Precision (kuning), F2-Score (biru), dan Akurasi (merah) terhadap variasi threshold probabilitas pada model. 

Secara umum, pola yang ditunjukkan oleh masing-masing metrik sesuai dengan ekspektasi:

- Recall (hijau) menurun seiring meningkatnya threshold, karena kriteria klasifikasi menjadi lebih ketat dan menyebabkan lebih banyak kasus positif yang terlewat.

- Precision (kuning) meningkat dengan naiknya threshold, menandakan bahwa prediksi positif semakin akurat meskipun dengan risiko menurunnya coverage terhadap kasus positif aktual.

- F2-Score (biru) menunjukkan penurunan bertahap, mencerminkan bahwa penurunan recall lebih dominan dibanding peningkatan precision. Tidak terdapat puncak yang tajam, namun nilai F2-score cenderung lebih tinggi pada threshold rendah. Dalam konteks ini, F2-score menjadi acuan utama untuk mengukur performa model dalam mendeteksi kasus positif secara agresif, selama precision masih dalam batas yang dapat diterima.

- Akurasi (merah) tampak relatif stabil atau menunjukkan peningkatan yang bersifat marginal seiring dengan perubahan threshold.

Threshold terpilih berada pada nilai **0.4**, dengan pertimbangan bahwa titik ini menghasilkan F2-score yang relatif tinggi, serta didukung oleh nilai recall dan precision yang seimbang sesuai dengan bobot F2 (yang lebih menekankan recall). Selain itu, akurasi yang dicapai pada threshold ini tetap kompetitif, sehingga memberikan keseimbangan yang baik antara deteksi kasus positif dan ketepatan prediksi secara keseluruhan.
